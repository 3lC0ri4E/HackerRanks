
# Extra Long Factorials Problem

Given an integer `n`, calculate and print the factorial of `n`. The factorial of `n` is defined as the product of all positive integers less than or equal to `n`.

### Input Format:

A single integer `n`.

### Constraints:
- `1 ≤ n ≤ 100`

### Output Format:

Print the factorial of `n`.

### Sample Input:

```
25
```

### Sample Output:

```
155112100433309859840000000
```

### Explanation:

The factorial of 25 is `25! = 25 * 24 * 23 * ... * 1 = 155112100433309859840000000`.

### Solution Explanation:

1. **Factorial Calculation**: The factorial of `n` is the product of all positive integers from `1` to `n`.
2. **Handling Large Numbers**: Since `n` can go up to `100`, the result may be a very large number. Python’s built-in integer type can handle arbitrarily large numbers, so there is no need to worry about overflow.
3. **Efficient Computation**: Factorial can be computed either iteratively or recursively.

---

## Python Solution:

```python
import math

def extraLongFactorials(n):
    # Calculate the factorial using math.factorial function
    return math.factorial(n)

if __name__ == "__main__":
    n = int(input())
    result = extraLongFactorials(n)
    print(result)
```

### Explanation of the Code:

1. **Importing the math Library**: Python's `math` module provides the `factorial()` function, which computes the factorial of a number.
2. **Calculating the Factorial**: The function `extraLongFactorials()` takes `n` as input and computes the factorial using the `math.factorial()` function.
3. **Input and Output**: The code reads an integer `n` from the input, computes the factorial, and prints the result.

---

### Time Complexity:

The time complexity is `O(n)`, as calculating the factorial involves multiplying `n` numbers together.

---

### Space Complexity:

The space complexity is `O(1)`, as we are using a single variable to store the result.

---

This solution efficiently computes the factorial of large numbers and handles them correctly.


````python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'extraLongFactorials' function below.
#
# The function accepts INTEGER n as parameter.
#

def extraLongFactorials(n):
    # Write your code here
    def factorial(n):
        if n == 0:
            return 1
        else:
            return n * factorial(n - 1)
    
    print(factorial(n))

if __name__ == '__main__':
    # Simulación de entrada para probar el código
    n = 25  # Número para calcular el factorial
    
    extraLongFactorials(n)
````
