
# Subarray Division

Two children, Lily and Ron, want to share a chocolate bar. Each of the squares has an integer on it. 

Lily decides to share a contiguous segment of the bar selected such that: 

  * The length of the segment matches Ron's birth month, and,
  * The sum of the integers on the squares is equal to his birth day.

Determine how many ways she can divide the chocolate.

## Function Description

Complete the `birthday` function in the editor below. 

The function has the following parameters:
  * **int s[n]**: the numbers on each of the squares of chocolate 
  * **int d**: Ron's birth day 
  * **int m**: Ron's birth month 

The function should return the number of ways the chocolate bar can be divided.

## Input Format
- The first line contains an integer `n`, the number of squares in the chocolate bar.
- The second line contains `n` space-separated integers, the numbers on the chocolate squares where `1 ≤ s[i] ≤ 5`.
- The third line contains two space-separated integers, `d` and `m`, Ron's birth day and his birth month.

## Constraints
- `1 ≤ n ≤ 100`
- `1 ≤ d ≤ 100`
- `1 ≤ m ≤ 12`

## Sample Input 0:
```
5
1 2 1 3 2
3 2
```

## Sample Output 0:
```
2
```

## Explanation 0:
Lily wants to give Ron 3 squares summing to 2. The following two segments meet the criteria:
- [1, 2, 1]
- [2, 1, 3]

## Sample Input 1:
```
6
1 1 1 1 1 1
3 2
```

## Sample Output 1:
```
0
```

## Explanation 1:
Lily only wants to give Ron 3 consecutive squares of chocolate whose integers sum to 2. There are no possible pieces satisfying these constraints.

## Sample Input 2:
```
1
4
4 1
```

## Sample Output 2:
```
1
```

## Explanation 2:
Lily only wants to give Ron 1 square of chocolate with an integer value of 4. Because the only square of chocolate in the bar satisfies this constraint, we print 1.


````python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'birthday' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER_ARRAY s
#  2. INTEGER d
#  3. INTEGER m
#

def birthday(s, d, m):
    # Write your code here
    substrings = []
    counter = 0
    for i in range(0, len(s) - m + 1):  # Only iterate up to len(s) - m to get valid subarrays
        substrings.append(s[i:m+i])
    for substring in substrings:
        if sum(substring) == d and len(substring) == m:
            counter += 1
    return counter

if __name__ == '__main__':
    # Simulación de entrada para probar el código
    s = [1, 2, 1, 3, 2]  # Lista de números
    d = 3  # Suma que buscamos en las sublistas
    m = 2  # Longitud de las sublistas

    result = birthday(s, d, m)

    # Mostrar el resultado de la prueba simulada
    print(result)
````
